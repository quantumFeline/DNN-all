```python
class SyntheticData[T](Dataset):
    """A small synthetic segmentation dataset.

    It is a sequence dataset of 5000 tuples (image, class, mask), where:
    - image: before transformation, an RGB PIL Image.
    - class: int 0..4, the label index.
    - mask:  numpy array of dtype=bool, shape (H, W), same size as image.
    """

    CLASSES: Final[tuple[str, ...]] = ("circle", "square", "triangle", "star", "diamond")

    def __init__(
        self,
        path: Path,
        split: Literal["train", "val", "test"],
        transform: Callable[[PIL.Image.Image], T],
    ) -> None:
        self.path = path
        self.transform = transform
        self.split = split

        self.class_to_idx = {class_name: i for i, class_name in enumerate(self.CLASSES)}

        images = sorted((self.path / "images" / self.split).glob("*.png"))
        masks = sorted((self.path / "masks" / self.split).glob("*.png"))
        labels = sorted((self.path / "labels" / self.split).glob("*.txt"))

        assert images, f"No images found in {self.path / 'images' / self.split}"
        assert len(images) == len(masks) == len(labels), (
            "Number of images, masks, and labels must be the same"
        )
        assert [p.stem for p in images] == [p.stem for p in masks] == [p.stem for p in labels], (
            "Image/mask/label filename mismatch."
        )
        self.image_names = [p.stem for p in images]

    def __len__(self) -> int:
        return len(self.image_names)

    def __getitem__(self, idx) -> tuple[T, int, np.ndarray]:
        name = self.image_names[idx]
        img_path = self.path / "images" / self.split / f"{name}.png"
        mask_path = self.path / "masks" / self.split / f"{name}.png"
        label_path = self.path / "labels" / self.split / f"{name}.txt"

        img = PIL.Image.open(img_path).convert("RGB")
        mask = np.array(PIL.Image.open(mask_path), dtype=bool)
        label = self.class_to_idx[label_path.read_text().strip()]

        img_transformed: T = self.transform(img)

        return img_transformed, label, mask


def show_image_row(
    image_dict: dict[str, PIL.Image.Image | np.ndarray | Tensor], size: float = 3.0
) -> None:
    n = len(image_dict)
    _, axs = plt.subplots(1, n, figsize=(size * n, size), constrained_layout=True, squeeze=True)
    if n == 1:
        axs = [axs]
    for ax, (title, img) in zip(axs, image_dict.items(), strict=True):
        ax.imshow(img)
        ax.set_title(title)
        ax.axis("off")
    plt.show()
```

## Training a classifier [do not modify]

```python
def train(
    model: nn.Module,
    train_loader: torch.utils.data.DataLoader,
    val_loader: torch.utils.data.DataLoader,
    test_loader: torch.utils.data.DataLoader,
    epochs: int = 5,
    **optimizer_kwargs: Any,
) -> None:
    print(f"ðŸš€ Training CNN for {epochs} epochs...")
    optimizer = optim.AdamW(model.parameters(), **optimizer_kwargs)

    for epoch in range(epochs):
        _train_epoch(model, train_loader, optimizer, desc=f"Epoch {epoch + 1}/{epochs} training  ")
        val_metrics = evaluate(model, val_loader, desc=f"Epoch {epoch + 1}/{epochs} validation")
        print(
            f"Epoch {epoch + 1}/{epochs} â€” "
            + f"val loss: {val_metrics['loss']:.3f}, val acc: {val_metrics['accuracy']:.1%}"
        )

    test_metrics = evaluate(model, test_loader, desc="Test Evaluation")
    print(
        "âœ… Model training complete: "
        + f"Test loss: {test_metrics['loss']:.3f}, test acc: {test_metrics['accuracy']:.1%}"
    )


def _train_epoch(
    model: nn.Module,
    dataloader: torch.utils.data.DataLoader,
    optimizer: optim.Optimizer,
    desc: str,
) -> dict[str, float]:
    model.train()
    device = next(model.parameters()).device
    total_loss = 0
    total_correct = 0
    total_samples = 0
    progress_bar = tqdm(dataloader, desc=desc)
    for imgs, labels, _ in progress_bar:
        imgs, labels = imgs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = nn.CrossEntropyLoss()(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

        _, predicted = torch.max(outputs.data, dim=1)
        total_samples += labels.shape[0]
        total_correct += (predicted == labels).sum().item()

        progress_bar.set_postfix(
            train_loss=f"{total_loss / (total_samples / labels.shape[0]):.3f}",
            train_acc=f"{total_correct / total_samples:.1%}",
        )

    return {"loss": total_loss / len(dataloader), "accuracy": total_correct / total_samples}


def evaluate(
    model: nn.Module, dataloader: torch.utils.data.DataLoader, desc: str
) -> dict[str, float]:
    model.eval()
    device = next(model.parameters()).device
    total_loss = 0
    total_correct = 0
    total_samples = 0
    with torch.no_grad():
        for imgs, labels, _mask in tqdm(dataloader, desc=desc):
            imgs, labels = imgs.to(device), labels.to(device)
            outputs = model(imgs)
            loss = nn.CrossEntropyLoss()(outputs, labels)
            total_loss += loss.item()
            _, predicted = torch.max(outputs.data, dim=1)
            total_samples += labels.shape[0]
            total_correct += (predicted == labels).sum().item()

    return {"loss": total_loss / len(dataloader), "accuracy": total_correct / total_samples}


class DataloaderArgs(TypedDict, total=False):
    batch_size: int
    shuffle: bool
    num_workers: int
    pin_memory: bool
```

## GradCAM implementation (add your code)

```python
from functools import partial
import torch.nn.functional as F

class GradCAM:
    """
    Pure PyTorch implementation of Grad-CAM.

    Usage:
        >>> model = torchvision.models.resnet18(pretrained=True)
        >>> target_layers = [model.layer4[-1]]
        >>> grad_cam = GradCAM(model=model, target_layers=target_layers)
        >>> heatmap = grad_cam(input_tensor, targets=[target_class_id])
    """

    def __init__(self, model: nn.Module, target_layers: Iterable[nn.Module]) -> None:
      """
      Initialize GradCAM with a model and target layers.

      Args:
          model: The neural network model to explain.
          target_layers: Layers to compute Grad-CAM for. If multiple layers are
              provided, the resulting heatmaps will be averaged.
      """
      self.model = model
      self.target_layers = target_layers

      self.activations = [None] * len(target_layers)
      self.gradients = [None] * len(target_layers)

      self.forward_hooks = [] # activation hooks
      self.backward_hooks = [] # gradient hooks

      # Register hooks for each target layer
      for i, layer in enumerate(self.target_layers):
          act_hook_fn = partial(self.save_activation, i)
          self.forward_hooks.append(layer.register_forward_hook(act_hook_fn))
          grad_hook_fn = partial(self.save_gradient, i)
          self.backward_hooks.append(layer.register_full_backward_hook(grad_hook_fn))


    def save_activation(self, layer_idx: int, _module: nn.Module,
                         _input: tuple, act_output: Tensor) -> None:
        """
        Forward hook to capture layer activations.

        Args:
            layer_idx: Index of the target layer.
            _module: The layer module (unused).
            _input: Tuple of inputs to the layer (unused).
            act_output: The layer's output tensor.
        """
        self.activations[layer_idx] = act_output.detach()

    def save_gradient(self, layer_idx, _module, _grad_input, grad_output):
        """
        Backward hook to capture gradients with respect to layer outputs.

        Args:
            layer_idx: Index of the target layer.
            _module: The layer module (unused).
            _grad_input: Tuple of gradients w.r.t. the layer's inputs (unused).
            grad_output: Tuple of gradients w.r.t. the layer's outputs.
        """
        self.gradients[layer_idx] = grad_output[0].detach()

    def clear_saved_data(self):
        """
        Clears the saved activations and gradients.
        """
        self.activations = [None] * len(self.target_layers)
        self.gradients = [None] * len(self.target_layers)

    def __call__(self, images: Tensor, targets: Iterable[int] | None = None) -> np.ndarray:
        """
        Compute Grad-CAM for the given input tensor.

        Args:
            images: input tensor to model (a normalized batch of shape B,C,H,W).
            targets: an iterable of target classes that we want to segment;
            if None is given, use the top class predicted by the model.

        Returns: numpy array of shape (B, H, W) containing the GradCam heatmap,
        values 0..1 (rescaled independently for each image in the batch
        and each target_layers)
        where
        B = batch size
        H = image height
        W = image width
        """
        target_classes = targets # readability alias
        batch_size = images.shape[0]
        height = images.shape[2]
        width = images.shape[3]

        # Forward propagation
        y = self.model(images)

        if target_classes is None:
            target_classes = torch.argmax(y, dim=1)

        # Backward propagation
        self.model.zero_grad()
        loss = y[range(batch_size), target_classes].sum()
        loss.backward(retain_graph=True)

        activations = self.activations
        gradients = self.gradients

        heatmaps = self.compute_heatmaps(images, activations, gradients)
        self.clear_saved_data()

        return heatmaps

    def compute_heatmaps(self, images: Tensor, activations: list[Tensor],
                         gradients: list[Tensor]) -> np.ndarray:
        """
        Compute Grad-CAM heatmaps from activations and gradients.

        Implementation follows the Grad-CAM paper:
        1. Compute importance weights as global average pooling of gradients
        2. Weighted combination of forward activation maps
        3. Apply ReLU to focus on features with positive influence
        4. Upsample to input resolution
        5. Normalize to [0, 1] range

        Args:
            images: Input tensor, shape (B, C, H, W).
            activations: List of activation tensors for each target layer,
                each of shape (B, C_layer, H_layer, W_layer).
            gradients: List of gradient tensors for each target layer,
                each of shape (B, C_layer, H_layer, W_layer).

        Returns:
            Numpy array of shape (B, H, W) with normalized heatmaps.
        """
        batch_size, _, height, width = images.shape

        all_layer_heatmaps = []

        for layer_i in range(len(activations)):
            activation = activations[layer_i]  # (B, C, H_layer, W_layer)
            gradient = gradients[layer_i]      # (B, C, H_layer, W_layer)

            weights = gradient.mean(dim=(2, 3)) # (B, C)

            # for each batch b and spatial position (h,w),
            # sum over channels c: weights[b,c] * activation[b,c,h,w]"
            heatmap = torch.einsum('bc,bchw->bhw', weights, activation)
            # (B, H_layer, W_layer)

            heatmap = F.relu(heatmap)

            # Upsample all images at once
            heatmap = F.interpolate(
                heatmap.unsqueeze(1),  # (B, 1, H_layer, W_layer)
                size=(height, width),
                mode='bilinear',
                align_corners=False
            ).squeeze(1)  # (B, H, W)

            heatmap = heatmap / (heatmap.amax(dim=(1, 2), keepdim=True) + 1e-8)

            all_layer_heatmaps.append(heatmap)

        all_layer_heatmaps = torch.stack(all_layer_heatmaps, dim=0)
        # (L, B, H, W)
        final_heatmaps = all_layer_heatmaps.mean(dim=0)
        # (B, H, W)

        return final_heatmaps.cpu().numpy()
```

```python
def heatmap_to_rgb_image(
    heatmap: np.ndarray, min: float | None = None, max: float | None = None
) -> PIL.Image.Image:
    """
    Converts a single-channel heatmap to an RGB pillow image using a colormap.

    Args:
    - heatmap: shape (H, W), will be normalized by mapping min..max to 0..1.
    - min: minimum value for normalization, defaults to heatmap.min().
    - max: maximum value for normalization, defaults to heatmap.max()
    """
    heatmap = heatmap.astype(np.float32)
    if min is None:
        min = heatmap.min()
    if max is None:
        max = heatmap.max()
    heatmap = (heatmap - min) / (max - min + 1e-8)
    heatmap_uint8 = (np.clip(heatmap, 0.0, 1.0) * 255).astype(np.uint8)
    heatmap_color = cv2.applyColorMap(heatmap_uint8, cv2.COLORMAP_JET)
    heatmap_color = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)
    return PIL.Image.fromarray(heatmap_color)
```

```python
class BasicSamPipeline:
    def __call__(self, images: Tensor) -> Tensor:
        """
        Input: normalized images, shape (B, C=3, H, W).
        Output: masks tensor of shape (B, H, W), dtype=bool.
        """
        B, C, H, W = images.shape
        # The basic pipeline always uses a single center point for each image.
        point_coords = np.array([[(W // 2, H // 2)] for _ in range(B)])

        # The basic pipeline always uses a single foreground point, no background points.
        point_labels = np.array([[1] for _ in range(B)], dtype=np.int64)

        return self.segment(images, point_coords, point_labels)

    def segment(
        self, images: Tensor, point_coords: np.ndarray, point_labels: np.ndarray
    ) -> Tensor:
        """
        Args:
        - images: normalized images, shape (B, C=3, H, W).
        - point_coords: point coordinates within each image, shape (B, num_points, 2), format (x,y).
            Note the format is not (h,w)=(y,x), but (x,y)!
        - point_labels: point labels, shape (B, num_points), dtype int64.
            Label 1 is foreground (should be in mask), 0 is background (shouldn't be in mask).

        Returns: segmentation masks, shape (B, H, W), dtype=bool.
        """
        B, C, H, W = images.shape
        assert C == 3, f"Expected images.shape=(B, C=3, H, W), got: {images.shape}"
        num_points = point_coords.shape[1]
        assert point_coords.shape == (B, num_points, 2), f"Expected point_coords.shape=({B=}, num_points, 2), got: {point_coords.shape}"
        assert point_labels.shape == (B, num_points), f"Expected point_labels.shape=({B=}, num_points), got: {point_labels.shape}"

        results = list[Tensor]()
        for image, pt_coords, pt_labels in zip(images, point_coords, point_labels, strict=True):
            sam_predictor.set_image(np.array(inverse_transform(image)))
            masks, scores, _logits = sam_predictor.predict(
                point_coords=pt_coords, point_labels=pt_labels, multimask_output=True
            )
            best_mask = masks[np.argmax(scores)]
            results.append(torch.tensor(best_mask, dtype=torch.bool))
        return torch.stack(results)
```

```python
import tqdm

class SamPipelineForegroundOnly(BasicSamPipeline):
    """
    SAM pipeline that automatically finds foreground points using Grad-CAM.

    This pipeline uses a Grad-CAM heatmap to identify the most salient region
    in the image, then computes a weighted centroid of high-activation areas
    as the foreground point prompt for SAM.
    """
    def __init__(self, grad_cam):
        self.grad_cam = grad_cam

    def __call__(self, images: Tensor) -> Tensor:
        """
        Segment objects using automatically detected foreground points.

        Args:
            images: Normalized RGB images, shape (B, 3, H, W).

        Returns:
            Binary segmentation masks, shape (B, H, W), dtype bool.
        """
        B, C, H, W = images.shape
        point_coords = np.array([[self.find_foreground_point(image)] for image in images])
        point_labels = np.array([[1] for _ in range(B)], dtype=np.int64)

        return self.segment(images, point_coords, point_labels)

    @staticmethod
    def weighed_centroid(img, threshold = 0):
        """
        Compute weighted centroid of array values above threshold.

        The centroid is weighted by the array intensity, focusing on
        the most salient regions.

        Args:
            img: 2D array, shape (H, W).
            threshold: Minimum heatmap value to include in centroid calculation.

        Returns:
            Tuple of (x, y) coordinates of the weighted centroid.
        """
        img_cut = np.where(img > threshold, img, np.zeros_like(img))
        x_coords, y_coords = np.indices(img_cut.shape)
        intensity = np.sum(img_cut.sum())
        x_center = np.round(np.sum(x_coords * img_cut) / intensity).astype(np.int32)
        y_center = np.round(np.sum(y_coords * img_cut) / intensity).astype(np.int32)
        return (y_center, x_center)

    def find_foreground_point(self, img: Tensor) -> tuple[int, int]:
        """
        Find the foreground point using Grad-CAM weighted centroid.

        Args:
            img: Single normalized image, shape (C, H, W).

        Returns:
            Tuple of (x, y) coordinates of the foreground point.
        """
        cam_image = self.grad_cam(img.unsqueeze(0).to(device))
        heatmap_threshold = 0.6

        return SamPipelineForegroundOnly.weighed_centroid(cam_image.squeeze(), threshold=heatmap_threshold)

    def find_foreground_point_max(self, img): # Alternative method, doesn't work as well
        cam_image = self.grad_cam(img.unsqueeze(0).to(device))
        max_point = np.unravel_index(cam_image[0].argmax(), cam_image[0].shape)
        # Need to swap coords for SamPipeline
        return (max_point[1], max_point[0])

    def evaluate_pointfinding_hit_rate(self, images: Tensor, ground_truth_masks: Tensor, use_max_point: bool = False) -> Tensor:
        """
        Uses centroid by default; max point otherwise.
        Returns the hit rate.
        """
        count = 0
        print("N images:", len(images))
        for image, mask in tqdm.tqdm(zip(images, ground_truth_masks)):
            if use_max_point:
                point = self.find_foreground_point_max(image)
            else:
                point = self.find_foreground_point(image)
            count += (mask[point[1], point[0]] == 1)
        return count / len(images)


    def evaluate_pointfinding_distance(self, images: Tensor, ground_truth_masks: Tensor, use_max_point: bool = False) -> Tensor:
        """
        Uses centroid by default; max point otherwise.
        Returns average distance.
        """
        distance_sum = 0
        for image, mask in tqdm.tqdm(zip(images, ground_truth_masks)):
            if use_max_point:
                point = self.find_foreground_point_max(image)
            else:
                point = self.find_foreground_point(image)
            ground_truth_mask_center = np.array(SamPipelineForegroundOnly.weighed_centroid(mask))
            distance_sum += np.linalg.norm(ground_truth_mask_center - np.array(point))
        return distance_sum / len(images)
```

```python
class SamPipelineForegroundAndBackground(BasicSamPipeline):
    """
    SAM pipeline that automatically finds both foreground and background points.

    Uses Grad-CAM to identify:
    - Foreground point: weighted centroid of high-activation region
    - Background point: location with minimum activation (likely background)
    """
    def __init__(self, grad_cam):
        self.grad_cam = grad_cam

    def __call__(self, images: Tensor) -> Tensor:
        """
        Input: normalized images, shape (B, C=3, H, W).
        Output: masks tensor of shape (B, H, W), dtype=bool.
        """
        B, C, H, W = images.shape
        point_coords = np.array([self.find_foreground_and_background_points(image) for image in images])
        point_labels = np.array([[1, 0] for _ in range(B)], dtype=np.int64)

        return self.segment(images, point_coords, point_labels)

    def find_foreground_and_background_points(self, img):
        """
        Find both foreground and background points using Grad-CAM.

        Args:
            image: Single normalized image, shape (C, H, W).

        Returns:
            List of two (y, x) tuples: [foreground_point, background_point].
        """
        cam_image = self.grad_cam(img.unsqueeze(0).to(device))

        heatmap_threshold = 0.6
        cam_image_cut = np.where(cam_image > heatmap_threshold, cam_image, np.zeros_like(cam_image))[0]
        x_coords, y_coords = np.indices(cam_image_cut.shape)
        intensity = np.sum(cam_image_cut.sum())
        x_center = np.round(np.sum(x_coords * cam_image_cut) / intensity).astype(np.int32)
        y_center = np.round(np.sum(y_coords * cam_image_cut) / intensity).astype(np.int32)

        # For minimum point, we can use a simpler approach
        min_point = np.unravel_index(cam_image[0].argmin(), cam_image[0].shape)
        # Need to swap coords for SamPipeline
        return [(y_center, x_center), (min_point[1], min_point[0])]
```

```python
def iou(masks_1, masks_2):
    """
    Returns IoU (intersection over Union) given two batches of masks of the same length.
    """
    assert masks_1.shape == masks_2.shape
    assert len(masks_1.shape) == 3
    intersection = torch.logical_and(masks_1, masks_2)
    union = torch.logical_or(masks_1, masks_2)
    return torch.sum(intersection, axis=(1,2)) / torch.sum(union, axis=(1,2))
```

```python
def test_pipeline(pipeline, test_loader, indices = None, progress_bar = False):

    if indices is not None:
        images, labels, gt_masks = next(iter(test_loader))
        images_selected, labels_selected, gt_masks_selected = images[indices], labels[indices], gt_masks[indices]
        results = pipeline(images_selected).cpu()
        for image, result, gt_mask in zip(images_selected, results, gt_masks_selected, strict=True):
            show_image_row(
                {
                    "Input image": inverse_transform(image),
                    "Basic SAM predicted mask": result,
                    "Ground-truth mask": gt_mask,
                }
            )

    else: # Full evaluation
        print("Evaluating on full test set...")
        if progress_bar: # Less batching, but you can track progress.
            print("Evaluating on full test set...")
            iou_results = []
            for imgs, labels, gt_masks in test_loader:
                batch_size = 32  # adjust based on memory
                num_batches = (len(imgs) + batch_size - 1) // batch_size

                for i in tqdm.tqdm(range(num_batches), desc="Processing batches"):
                    start_idx = i * batch_size
                    end_idx = min(start_idx + batch_size, len(imgs))

                    batch_imgs = imgs[start_idx:end_idx]
                    batch_masks = gt_masks[start_idx:end_idx]

                    results = pipeline(batch_imgs.to(device)).cpu()
                    iou_results.append(iou(results, batch_masks))

            iou_results = torch.cat(iou_results).cpu().numpy()
            print("Average IoU:", np.mean(iou_results))
        else:
            images, labels, gt_masks = next(iter(test_loader))
            # batch size == test size, so we don't iterate
            results = pipeline(images.to(device)).cpu()
            iou_results = iou(results, gt_masks)

        print("Average IoU:", np.mean(iou_results))
```
